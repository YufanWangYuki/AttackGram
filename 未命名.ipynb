```python
import os
import pdb

import Levenshtein
```

```python
filePath = '/Users/yufanwang/Desktop/Study/Project/dataset/nearest/'
files = sorted(os.listdir(filePath))[1:]
files
```

```python
pred = files[2]
src = files[5]
tgt = files[8]
```

```python
f = open(filePath + '/' + pred)
lines = f.readlines()
preds = []
for line in lines:
    words = line.strip().replace("\n","").split(" ")
    preds.append((" ").join(words).lower())
```

```python
f = open(filePath + '/' + src)
lines = f.readlines()
srcs = []
for line in lines:
    words = line.strip().replace("\n","").split(" ")
    srcs.append((" ").join(words).lower())
```

```python
diff = {}
dists = []
diff_cnt = 0
for src, pred in zip(srcs, preds):
    if src != pred:
        diff_cnt += 1
    dist = Levenshtein.distance(src,pred)
    dists.append(dist)
    src_list = src.split(" ")
    pred_list = pred.split(" ")
    for word_src, word_pred in zip(src_list,pred_list):
        if word_src != word_pred:
#             if word_src == "july" and word_pred == 'august':
#                 pdb.set_trace()
            if (word_src.lower(), word_pred.lower()) not in diff:
                diff[(word_src.lower(), word_pred.lower())] = 1
            else:
                diff[(word_src.lower(), word_pred.lower())] += 1
```
