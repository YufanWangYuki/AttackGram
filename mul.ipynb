```python
with open('/Users/yufanwang/Desktop/Study/Project/dataset/nearest/tokenId_2_embed.pkl', 'rb') as f:
    n = pickle.load(f)
```

```python
res = []
means = []
for key,item in n.items():
    norm = np.linalg.norm(item,ord=1)/768
    res.append(norm)
```

```python
fig, ax = plt.subplots(dpi=600,figsize=(6,4))
plt.hist(res)
ax.set_xlabel('$\\mathit{l_1}\: norm$') #x x label
ax.set_ylabel('Count') #y y label
```
